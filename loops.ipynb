## for loop

```python
my_list = (1,2,3,4,5)
for box in my_list:
    
    print(box)
```

```python
name = "amaan sajid"
for char in name:
    print(char,end=" ")
```

```python
# 1,10
for i in range(1,11,2): # 1 --> start,10-->stop(n-1),2-->step
    print(i)
```

```python
for n in range(2,21,2):
    print(n)
```

#### create a table in for loop and take input from user

```python
num = int(input('Enter Your Number: '))
for i in range(1,11):
    print(f"{num}x{i}={i*num}")
```

```python
for i in range(1,11):
    if i %2 == 0:
        print(f"{i}: even")
    else:
        print(f'{i}: odd')
```

```python
num = 22
if num % 2 == 0:
    print('even')
else:
    print('odd')
```

### jumping statments
* break ---> stop execution
* continue ---> iteration skip 
* pass ----> as a placeholder

```python
for n in range(1,11):
    if n == 5:
        break

    print(n)
```

```python
name
```

```python
for n in name:
    if n == " ":
        break
    print(n)
```

```python
for n in name:
    if n == " ":
        continue
    print(n)
```

```python
for n in range(1,11):
    if n == 5:
        continue

    print(n)
```

```python
# pass 
for i in range(1,12):
    pass
```
