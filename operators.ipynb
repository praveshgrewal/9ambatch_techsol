# operatros (some operations)

### arthmetic operators

```python
a = 10
b = 20
print(a+b) # addition
print(a-b) # sub
print(a*b) # multip
print(a/b) # divison ---> decimal
print(a//b) # floor divison ---> whole number
print(a%b) # reminder
print(a**b) # power
```

```python
print(2**4)
```

### assinment

```python
num = 23 # equal 
print(num)
num = num+2

print(num)
```

```python
num += 2 # num = num+2
print(num)
```

```python
num -= 2  # num = num-2
print(num)
```

### comparision (relationl operators)

```python
print(10 == 10 ) # equal to 
print(10 != 10 ) # not equal to 
```

```python
num2 = 12
num3 = 12
print(num2 == num3)
```

```python
print(12 > 12) # greater then
print(12 >= 12) # greater then equal to 
print(12 < 34) # less then
print(12 <= 34) # less then equal to 
```

### membership operators ---> in , not in

```python
name = "sajid khan"
print('s' in name)
print('s' not in name)
```

### logical operators

```python
print(True and True) # we need all condition true
print(True and False)
print(True or False) # we need only one condition true
print(not True)
print(not False)
```

```python
print((10>1) and(11>1) )
```

### identity operators --> is

```python
x = 10
y = 10
z = 12
print(id(x))
print(id(y))
print(id(z))
print(x is y)
print(x is not z)
```
